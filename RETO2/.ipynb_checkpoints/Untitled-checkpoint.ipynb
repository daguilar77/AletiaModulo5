```python
# Algunas bibliotecas necesarias
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import sklearn
```

```python
# Para los primeros ejercicios vamos a crear 4 grupos de datos en 2 dimensiones.
```

```python
from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
# Creamos una lista de la etiqueta del grupo al que pertenecen los datos
#
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
print(y_kmeans)
```

```python
# Coloreamos los datos con esas etiquetas
# Creamos los centroides. GrafiquÃ©moslos igualmente
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
# Para encontrar los grupos, sklearn utiliza un algoritmo llamado Expectation maximization (E-M)
# Inventa aleatoriamente los centroides
# Repite los siguientes pasos hasta hallar la convergencia:
# Paso-E: Asigna cada uno de los ejemplos al centroide mas cercano formando un grupo para cada centroide
# Paso-M: Re calcula los centroides como el promedio de los puntos asignados a su grupo
```

```python
from sklearn.cluster import KMeans
import numpy as np

def mi_algoritmo_kmeans(X, n_clusters, semilla):
    # Inicializa el algoritmo KMeans
    kmeans = KMeans(n_clusters=n_clusters, random_state=semilla)
    kmeans.fit(X)
    
    # Obtiene los centroides y las etiquetas
    centroides = kmeans.cluster_centers_
    etiquetas = kmeans.labels_
    
    return centroides, etiquetas
```

import random
import numpy as np

def mi_algoritmo_kmeans(X, n_clusters, semilla):
    # Inicializa los centroides de manera aleatoria
    random.seed(semilla)
    centroides = X[random.sample(range(len(X)), n_clusters)]
    
    # Repite hasta que la convergencia se alcance
    while True:
        # Crea las etiquetas asignando cada punto al centroide mÃ¡s cercano
        etiquetas = np.array([np.argmin([np.linalg.norm(x-c) for c in centroides]) for x in X])
        
        # Calcula los nuevos centroides como la media de los puntos en cada cluster
        nuevos_centroides = [X[etiquetas == k].mean(axis=0) for k in range(n_clusters)]
        
        # Verifica si los centroides han convergido
        if np.allclose(centroides, nuevos_centroides):
            break
        else:
            centroides = nuevos_centroides
            
    return np.array(centroides), etiquetas

```python
import numpy as np
import matplotlib.pyplot as plt

def mi_algoritmo_kmeans(X, n_clusters, semilla):
    np.random.seed(semilla)
    
    # Inicializa los centroides aleatoriamente
    centroides = X[np.random.choice(X.shape[0], n_clusters, replace=False), :]
    
    # Calcula la distancia euclidiana entre los puntos y los centroides
    def distancia_euclidiana(a, b):
        return np.sqrt(np.sum((a - b)**2))
    
    # Repite hasta que los centroides no cambien
    etiquetas_previas = np.zeros(X.shape[0])
    while True:
        # Asigna cada punto al centroide mÃ¡s cercano
        etiquetas = np.array([np.argmin([distancia_euclidiana(x, centroides[i]) for i in range(n_clusters)]) for x in X])
        
        # Si no hubo cambios en las etiquetas, termina
        if np.array_equal(etiquetas, etiquetas_previas):
            break
        
        # Actualiza las etiquetas previas
        etiquetas_previas = np.copy(etiquetas)
        
        # Recalcula los centroides como la media de los puntos en cada cluster
        for i in range(n_clusters):
            centroides[i] = np.mean(X[etiquetas == i], axis=0)
    
    return centroides, etiquetas

# Crea una variable multidimensional con datos aleatorios
X = np.random.rand(300, 2)

# Ejecuta la funciÃ³n "mi_algoritmo_kmeans" con 4 clusters y semilla 123
centroides, etiquetas = mi_algoritmo_kmeans(X, 4, 123)

# Grafica los resultados
plt.scatter(X[:, 0], X[:, 1], c=etiquetas)
plt.scatter(centroides[:, 0], centroides[:, 1], marker='x', s=200, linewidths=3, color='r')
plt.show()
```
