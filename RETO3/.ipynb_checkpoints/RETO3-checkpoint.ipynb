```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

# Reto sobre overfitting & underfitting
Aproximar la función:
Y=cos(1.5πX)  - - - - - - - - - - - - - - - -  ecuación1 
utilizando una regresión lineal y con solo 30 observaciones de esta función
Pregunta 1: ¿Cuales son las principales complicaciones de este planteamiento?

```python
# Generar 30 observaciones para X
X = np.linspace(0, 1, 30).reshape(-1, 1)

# Calcular Y con la función cos(1.5πX)
Y = np.cos(1.5 * np.pi * X).reshape(-1, 1)

# Crear un modelo de regresión lineal
reg = LinearRegression().fit(X, Y)

# Calcular la aproximación con el modelo
Y_pred = reg.predict(X)

# Gráfico de comparación
plt.plot(X, Y, 'o', label='real')
plt.plot(X, Y_pred, '-', label='aproximación')
plt.legend()
plt.show()
```

# Respuesta pregunta 1:
### Aproximar una función sinusoidal con una regresión lineal puede ser un reto debido a la naturaleza no lineal de la función. La regresión lineal trabaja mejor con relaciones lineales entre las variables, mientras que la función seno es una función cíclica y, por lo tanto, no lineal.

### Además, utilizar solo 30 observaciones puede resultar en una falta de precisión en la aproximación, ya que la función seno puede tener muchas variaciones y cambios bruscos en un pequeño intervalo.

### Una posible solución a estos problemas sería utilizar técnicas de procesamiento no lineal, como la regresión polinómica o las redes neuronales, o utilizar más observaciones para tener una mejor representación de la función.

```python
# Generemos 30 datos a partir de la función planteada en la ecuación 1
# Adicionalmente para complicarle un poco el trabajo al modelo agreguemos un poco de ruido

np.random.seed(0)
n_samples = 30

def true_fun(X):
    return np.cos(1.5 * np.pi * X)

X = np.sort(np.random.rand(n_samples))
y_sin_ruido = true_fun(X)
y = true_fun(X) + np.random.randn(n_samples) * 0.1

plt.plot(X, label='X')
plt.plot(y_sin_ruido, label='y (muestra de función original)')
plt.plot(y, label='y (muestra con ruido)')
plt.legend();
```

# Pregunta 2
Recordando el planteamiento de la regresión lineal, intentaremos estimar  y  a partir de  X , es decir:  y^=θTX 
Sin embargo la respuesta de la pregunta 1 nos lleva a la conclusión de que no es posible estimar  y  a partir de  x  con una regresión lineal.
Pero en el análisis de datos existen algunos artilugios matemáticos para solucionar este tipo de problemas.
Lo que haremos es cambiar el planteamiento. Además de usar  X  usaremos algunas transformaciones no lineales de  X  y las usaremos como features adicionales.
Por ejemplo:  X2 ,  X3 ,  X4 ,  ... ,  Xn  (transformaciones polinomiales)
Los ejemplos anteriores son transformaciones polinomiales de grado 2, 3, 4, ..., n
Y esto nos lleva a...
Pregunta 2: ¿Cuántos grados son necesarios?

# Respuesta 2
### La cantidad de grados necesarios en una regresión polinómica para aproximar una función sinusoidal depende de la complejidad de la función a aproximar y del grado de precisión requerido. En general, es recomendable empezar con un polinomio de grado bajo y aumentar su grado hasta encontrar una aproximación adecuada. 

### El error cuadrático medio (MSE) puede ser utilizado para determinar la cantidad de grados necesarios en una regresión polinómica. Para hacer esto, se ajusta una regresión polinómica de grado creciente y se calcula el MSE de cada modelo. Luego, se selecciona el modelo con el MSE más bajo. Este modelo es el que mejor se ajusta a los datos y es probable que sea el que requiera la cantidad adecuada de grados.

### Se puede comenzar con un polinomio de grado 1, calcular el MSE, luego aumentar el grado a 2 y calcular el MSE nuevamente, y continuar este proceso hasta encontrar el modelo con el MSE más bajo. Este modelo sería el que mejor se ajusta a los datos y probablemente requerirá la cantidad adecuada de grados para aproximar la función sinusoidal.

### Es importante tener en cuenta que a medida que se aumenta el grado del polinomio, el modelo se vuelve más complejo y puede sufrir de overfitting, es decir, ajustarse excesivamente a los datos de entrenamiento y no generalizar bien a datos nuevos. Por lo tanto, es importante monitorear el MSE tanto en el conjunto de entrenamiento como en un conjunto de prueba separado para evitar overfitting y elegir el modelo que tenga un buen equilibrio entre precisión y complejidad.

A continuación te mostramos una manera de usar pipelines para explorar hiperparámetros.

Los pipelines son muy útiles al momento de explorar hiperparámetros (en este caso el hiperparámetro que estamos explorando es el máximo grado de libertad en la transformación de la variable  X  necesario para estimar  y )

```python
degrees = [1, 4, 15]

plt.figure(figsize=(14, 5))
for i in range(len(degrees)):
    ax = plt.subplot(1, len(degrees), i + 1)
    plt.setp(ax, xticks=(), yticks=())

    polynomial_features = PolynomialFeatures(degree=degrees[i], include_bias=False)
    linear_regression = LinearRegression()
    pipeline = Pipeline([("polynomial_features", polynomial_features),("linear_regression", linear_regression),])
    pipeline.fit(X[:, np.newaxis], y)

    X_test = np.linspace(0, 1, 100)
    plt.plot(X_test, pipeline.predict(X_test[:, np.newaxis]), label="Modelo creado")
    plt.plot(X_test, true_fun(X_test), label="Función original")
    plt.scatter(X, y, edgecolor="b", s=20, label="Muestra con ruido")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim((0, 1))
    plt.ylim((-2, 2))
    plt.legend(loc="best")
    plt.title("Modelo usando desde X^1 hasta X^{}".format(degrees[i]))
plt.show()
```

En las 3 gráficas anteriores se hacen evidentes los conceptos de underfitting, correct fit y overfitting transformando  X  desde el grado 1 hasta los grados 1, 4, 15 respectivamente.

Estos números con el máximo grado fueron determinados al tanteo probando desde 1 hasta 15 y este es un procedimiento válido para encontrar el fit correcto.

Sin embargo tiene dos desventajas muy importantes:

Es necesario ver la gráfica para determinar si el fit es el correcto (esto no es posible cuando la dimensión de X es mayor a 1).
Pierde la posibilidad de ser automatizado.

Además es importante mirar no solo el error en el set de entrenamiento, sino también en el de prueba. A continuación te mostramos la tan famosa gráfica de bias vs variance:

Pregunta 3: ¿Como podemos modificar el código anterior para incluir el cálculo del error y sistematizar la selección del grado máximo polinomial en la transformación de  X ?

```python
# RESPUESTA PREGUNTA 3
from sklearn.metrics import mean_squared_error

degrees = [1, 4, 15]
MSE = []
plt.figure(figsize=(14, 5))
for i in range(len(degrees)):
    ax = plt.subplot(1, len(degrees), i + 1)
    plt.setp(ax, xticks=(), yticks=())

    polynomial_features = PolynomialFeatures(degree=degrees[i], include_bias=False)
    linear_regression = LinearRegression()
    pipeline = Pipeline([("polynomial_features", polynomial_features),("linear_regression", linear_regression),])
    pipeline.fit(X[:, np.newaxis], y)
    
    y_pred = pipeline.predict(X[:, np.newaxis])
    mse = mean_squared_error(y, y_pred)
    MSE.append(mse)

    X_test = np.linspace(0, 1, 100)
    plt.plot(X_test, pipeline.predict(X_test[:, np.newaxis]), label="Modelo creado")
    plt.plot(X_test, true_fun(X_test), label="Función original")
    plt.scatter(X, y, edgecolor="b", s=20, label="Muestra con ruido")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim((0, 1))
    plt.ylim((-2, 2))
    plt.legend(loc="best")
    plt.title("Modelo usando desde X^1 hasta X^{}".format(degrees[i]))
    

print("Los valores MSE para cada grado son: ", MSE)
best_degree = degrees[np.argmin(MSE)]
print("El grado máximo polinomial es:", best_degree)
plt.show()
```

### Respuesta= calculamos el error MSE para cada modelo y lo almacenamos en una lista MSE. Finalmente, utilizamos la función np.argmin para encontrar el grado máximo polinomial que tiene el error MSE más bajo y lo imprimimos como resultado.

### Es importante tener en cuenta que a medida que se aumenta el grado del polinomio, el modelo se vuelve más complejo y puede sufrir de overfitting, es decir, ajustarse excesivamente a los datos de entrenamiento y no generalizar bien a datos nuevos. Por lo tanto, es importante monitorear el MSE tanto en el conjunto de entrenamiento como en un conjunto de prueba separado para evitar overfitting y elegir el modelo que tenga un buen equilibrio entre precisión y complejidad.
 

El proceso podría ser el siguiente:

Dividir los datos en dos conjuntos, uno de entrenamiento y otro de validación.
Entrenar varios modelos polinómicos con diferentes grados, usando solo los datos de entrenamiento.
Calcular el MSE en ambos conjuntos para cada modelo polinómico.
Seleccionar el modelo que brinde el menor MSE en el conjunto de validación.
Este proceso previene el overfitting y el underfitting, ya que el modelo se evalúa sobre datos que no han sido utilizados en el entrenamiento, y por tanto, refleja mejor la capacidad del modelo para generalizar a nuevos datos.

# Pregunta 4: ¿Cuales son otros hiperparámetros que incrementan la complejidad de los modelos? en el caso de:
Regresión logística
Árboles de decisión
K-medias
Redes neuronales

# Respuesta 4

En Regresión Logística, los hiperparámetros que incrementan la complejidad incluyen el tipo de regularización (L1 o L2), el valor del coeficiente de regularización, y el número de iteraciones.

En Árboles de Decisión, los hiperparámetros que incrementan la complejidad incluyen la profundidad máxima del árbol, el número mínimo de muestras requerido en un nodo para realizar una partición, y el número mínimo de muestras requerido en un nodo hoja.

En K-medias, los hiperparámetros que incrementan la complejidad incluyen el número de clústeres (K) y la inicialización de los centroides.

En Redes Neuronales, los hiperparámetros que incrementan la complejidad incluyen el número de capas ocultas, el número de unidades en cada capa, la función de activación, el tipo de regularización, y el tamaño del lote.

### Ahora, preparemos los datos para un modelo

```python
# Note that of the 3 levels of RecentTrend, we only created dummies for rising and falling
# We will be incuding constant in our model (dummy variable trap)
```

```python
# Para crear variables dummy para "Recent Trend", debes tener una columna llamada "RecentTrend" en el DataFrame "fulldf"
fulldf.rename(columns={'recent_trend':'RecentTrend'}, inplace=True)

# change all the missing values to the mode ('stable')
fulldf.replace({'RecentTrend' : {'*':'stable'}}, inplace=True)

# function to do boolean check and return 1 or 0
def f(x, term):
    if x == term:
        return 1
    else:
        return 0

# create new features using the apply method with the 'f' function we defined above
fulldf['Rising'] = fulldf.RecentTrend.apply(lambda x: f(x, term='rising'))
fulldf['Falling'] = fulldf.RecentTrend.apply(lambda x: f(x, term='falling'))
```

Si recordamos, la función Incidence_Rate contiene algunos valores (p. ej., '*', '||' y '|') con los que debemos tratar antes de poder convertirlos a formato numérico. Hagamos eso ahora. Afortunadamente, pandas lo hace bastante fácil.

El argumento "errors='coerce'" del método pd.to_numeric le dice a pandas que reemplace cualquier valor que no se pueda convertir en un flotante con 'nan'.

```python
X['Incidence_Rate'] = pd.to_numeric(X.Incidence_Rate, errors='coerce')
```

Ahora podemos usar el método pandas fillna() para imputar fácilmente los valores faltantes. Si bien es tentador usar una técnica más compleja para imputar los valores faltantes (como la regresión aleatoria de bosques, por ejemplo), la imputación usando la media o la mediana es mucho más simple y es probable que arroje resultados bastante comparables.

```python
X['Incidence_Rate'] = X.Incidence_Rate.fillna(X.Incidence_Rate.median())
print(sum(X.Incidence_Rate.isnull()))
```

### multicolinealidad
Apliquemos factores de inflación de varianza para evaluar la multicolinealidad. Los VIF, al realizar una variable independiente en la matriz de diseño que comprende todas las demás variables independientes, nos permite evaluar el grado en que esa variable independiente es ortogonal a las demás. Los VIF más grandes indican multicolinealidad.
(!) Nota: los valores de VIF superiores a 10 se interpretan comúnmente como "altos" en multicolinealidad. Sin embargo, este umbral es arbitrario y puede tener sentido seleccionar un umbral diferente en muchos casos.

Avancemos y echemos un vistazo a los residuos para evaluar la normalidad y la heteroscedasticidad.

### ¿Están los residuos normalmente distribuidos?

```python
# histogram superimposed by normal curve
plt.figure(figsize=(10,6))
import scipy.stats as stats
mu = np.mean(result.resid)
sigma = np.std(result.resid)
pdf = stats.norm.pdf(sorted(result.resid), mu, sigma)
plt.hist(result.resid, bins=100)
plt.plot(sorted(result.resid), pdf, color='r', linewidth=2)
plt.show()
```

La distribución de los residuos no se adhiere perfectamente a una distribución normal (sesgo=0, exceso de curtosis=0). Hay un pequeño número de valores atípicos a la izquierda, las colas parecen ligeramente más gruesas y la distribución tiene una curtosis algo mayor que la distribución normal.

```python
# QQplot
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(10,3))
sm.qqplot(result.resid, stats.t, fit=True, line='45', ax = ax1)
ax1.set_title("t distribution")
sm.qqplot(result.resid, stats.norm, fit=True, line='45', ax=ax2)
ax2.set_title("normal distribution")
plt.show()
```

Los diagramas qq confirman que los residuos se adhieren más a la distribución t que a la distribución normal (colas más anchas). Algunos valores atípicos prominentes son visibles en el extremo inferior.

En definitiva, a pesar de estas imperfecciones, considero adecuada la distribución de residuos. Sin embargo, debemos investigar la naturaleza de los valores atípicos más extremos. También es posible que deseemos intentar agregar información adicional al modelo (nueva(s) variable(s)), a la luz de la tendencia del modelo a sobrestimar los valores bajos y subestimar los valores altos.

### Heterocedasticidad

```python
# plot predicted vs actual
plt.figure(figsize=(14,7))
sns.regplot(x=y, y=result.fittedvalues, line_kws={'color':'r', 'alpha':0.3, 
                                                  'linestyle':'--', 'linewidth':2}, 
            scatter_kws={'alpha':0.5})

plt.ylim(0,160)
plt.xlabel('Actual Values')
plt.ylabel('Fitted Values')
plt.show()
print("Pearson R: ", stats.pearsonr(result.fittedvalues, y))
```

De acuerdo con nuestro valor de R^2 informado, ahora visualizamos la fuerte correlación entre los valores reales y predichos.

```python
# plot actual values versus residuals
from statsmodels.nonparametric.smoothers_lowess import lowess
ys = lowess(result.resid.values, y, frac=0.2)
ys = pd.DataFrame(ys, index=range(len(ys)), columns=['a', 'b'])
ys = ys.sort_values(by='a')

fig, ax = plt.subplots(figsize=(14,9))
plt.scatter(y, result.resid, alpha=0.5, s=25)
plt.axhline(y=0, color='r', linestyle="--", alpha=0.5)
plt.xlabel("Actual Values")
plt.ylabel("Residuals")

plt.plot(ys.a, ys.b, c='green', linewidth=2, label="Lowess")
plt.legend()
plt.show()
print("Pearson R:", stats.pearsonr(y, result.resid))
```

Este gráfico sugiere que el modelo tiende a sobrepasar un poco los valores más bajos y por debajo de los más altos.

```python
# plot actual values versus residuals
plt.figure(figsize=(14,7))
plt.scatter(y=result.resid, x=result.fittedvalues, alpha=0.5, s=22)
plt.axhline(y=0, color='r', linestyle="--", alpha=0.5)
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.ylim(-65, 50)
plt.show()
```

La gráfica de residuos versus valores ajustados muestra que los residuos son bastante simétricos en relación con los valores ajustados.
